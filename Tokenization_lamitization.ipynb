# Stopword Removal

Stopwords are common words that do not add much meaning to a sentence, such as "the", "is", and "and". NLTK provides a stopword module that contains a list of stop words for various languages. We can use this module to remove stop words from our tokens.

```python
import nltk
nltk.download('stopwords')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

text = "Natuaral Language Processing is a subfield of artificial intelligence that focuses on the interaction between humans and computers using natural language."

tokens = word_tokenize(text)
stop_word = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_word]
print(filtered_tokens)
```

# Lemmatization

Lemmatization is the process of reducing words to their base form, or lemma. For example the word "running" would be reduced to "run". This can help to group together words with similar meanings. NLTK provides a WordNetLemmatizer class that performs lemmatization.

```python
import nltk
nltk.download('wordnet')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

text = "Natuaral Language Processing is a subfield of artificial intelligence that focuses on the interaction between humans and computers using natural language."
tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

lemmatizer = WordNetLemmatizer()
lemmatizer_tokens = [lemmatizer.lemmatize(tokens) for tokens in filtered_tokens]

print(lemmatizer_tokens)
```

# Sentiment Analysis

Sentiment analysis is the process of determining the emotional tone of a piece of text. NLTK provides a SentimentlntensitAnalyzer class that analyzes text for its negative, neutral, and positive sentiment.

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer

text = "I love natural language processing! It's so interesting and useful."

analyzer = SentimentIntensityAnalyzer()
sentiment_scores = analyzer.polarity_scores(text)

print(sentiment_scores)
```

# Named Entity Recognition

Named entity recognition (NER) is the process of identifying  and classifying named entities in text, such as people, organization, and locations.
NLTK provides a ne_chunk function that performs NER.

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
```

```python
import nltk
nltk.download('maxent_ne_chunker')
nltk.download('words')
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import ne_chunk

text = "Barack Obama was born in Hawaii and served and the 44th President of the United States from 2009 to 2017."

tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

lemmatizer = WordNetLemmatizer()
lemmatizer_tokens = [lemmatizer.lemmatize(tokens) for tokens in filtered_tokens]

pos_tags = nltk.pos_tag(lemmatizer_tokens)

named_entities = ne_chunk(pos_tags)

print(named_entities)
```
