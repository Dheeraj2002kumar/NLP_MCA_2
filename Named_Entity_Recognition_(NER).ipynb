```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
# input text
text = """
My name is Dheeraj Kumar, I belongs to Hazaribag, Jharkhand.
Master of Computer Application from Amity University, Mumbai.
"""
```

```python
# process the text
doc = nlp(text)
```

```python
print("------------- Named Entity Recognition (NER) output -------------\n")

# Extract and print named entities
for ent in doc.ents:
  print(f"Entity Text: {ent.text}")
  print(f"Entity Label: {ent.label_}")
  print(f"Entity Text: {spacy.explain(ent.label_)}")
```

```python
# Input text
text = """
Ratan Tata was the Chairman of Tata Group.
Elon Mask is the CEO of Tesla and SpaceX.
Infosys is headquatered in Bangalore.
"""

# Process the text
doc = nlp(text)

print("------------- Named Entity Recognition (NER) output -------------\n")

# Extract and print named entities
for ent in doc.ents:
  print(f"Entity Text: {ent.text}")
  print(f"Entity Label: {ent.label_}")
  print(f"Entity Text: {spacy.explain(ent.label_)}")
```
